# Utilities

```python
#Imports the libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Data 

```python
#Defines the values for each of the five epochs for the various batch sizes
GPU64 = [15.87,16.03,15.6,16.09,15.43]
GPU128 = [19.56, 12.44, 12.61,12.56,12.67]
GPU256 = [18.47, 11.81,11.97,12.05,12.02]
GPU512 = [11.36,11.32,11.28,11.27,11.21]
GPU1024 = [11.32,11.29,11.13,11.25,11.24]
```

```python
#Defines the values for each of the five epochs for the various batch sizes
CPU64 = [159.82,170.34,172.42,171.76,172.47]
CPU128 = [148.37, 154.09,155.4,155.37,155.69]
CPU256 = [184.68,179.36,180.68,190.11,182.42]
CPU512 = [196.21,175.1,186.65,176.78,178.89]
CPU1024 = [202.34, 208.97,204.3,200.68,216.68]
```

```python

# create sample data and takes the mean of each previous defined list
data = {'Batch_Size': [64, 64, 128, 128, 256, 256, 512, 512, 1024, 1024],
        'Time': [np.mean(CPU64),np.mean(GPU64), np.mean(CPU128),np.mean(GPU128),np.mean(CPU256),np.mean(GPU256), np.mean(CPU512),np.mean(GPU512), np.mean(CPU1024),np.mean(GPU1024)],
        'Type': ['CPU', 'GPU','CPU', 'GPU','CPU', 'GPU','CPU', 'GPU','CPU','GPU']}
df = pd.DataFrame(data) #Adds the data to the main dataframe

df_cpu = df[df['Type'] == 'CPU']
df_gpu = df[df['Type'] == 'GPU']
```

# Data Visualization 

```python
# Defines the subplot
fig, ax = plt.subplots()
ax.set_xlabel('Batch Size')
ax.set_ylabel('Time(in seconds)')

#Adds in log scale for the x-axis
ax.set_xscale('log')

#Plots the CPU and the GPU 
ax.plot(df_cpu['Batch_Size'], df_cpu['Time'], label='CPU',color='blue')
ax.plot(df_gpu['Batch_Size'], df_gpu['Time'], label='GPU', color='red')

#Adds the legend
ax.legend()

#Adds the title 
plt.title('CPU vs GPU Batch Size Comparison')

#Shows the plot 
plt.show()

#NOTE: Code and implementation of this graph is borrowed from CS: 5781 Machine Learning Engineering Module 3
```
